```python
import os
import math
import string
```

```python
Prefix = [i+i+i for i in string.ascii_lowercase]
```

```python
print(Prefix)
```

```python
root_dir = "/Users/weizhenliu/Downloads/AntsAndBees"
```

```python
group = [d for d in os.listdir(root_dir) if not d[0] == '.']
classes = [d for d in os.listdir(os.path.join(root_dir,group[0])) if not d[0] == '.']
# len(os.listdir(os.path.join(root_dir,group[0],classes[0])))
print(group,classes)
```

```python
for g in group:
    for c in classes:
        if c == "bees":
            m = 0
        if c == "ants":
            m = 12
#         num = math.ceil(len(os.listdir(os.path.join(root_dir,g,c)))/10)
#         for m in range(j,26,2):
# #             print(m)
        for i,f in enumerate(os.listdir(os.path.join(root_dir,g,c))):
            os.rename(os.path.join(root_dir,g,c,f),os.path.join(root_dir,g,c,Prefix[m]+"_"+str(i)+".jpg"))
            m += 1

            if c == 'bees' and m == 12:
                m = 0
            elif c == 'ants' and m == 26:
                m = 12
            print(g,c,m,Prefix[m])
#                 if (c == "val" and j == 10) or (c == 'train' and j == 26):
#                     break
      
```
