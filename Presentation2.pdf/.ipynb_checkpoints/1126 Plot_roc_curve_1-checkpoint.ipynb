```python
import random
def adjust_data(data, down_value=0.2, decrease=None):
    N = len(data)
    for ith, it in enumerate(data):
        if ith > N*5/6:
            continue
        it = it - down_value*(1-ith/N)
        if decrease is not None:
            it = it - decrease
        if it<0:
            it = 0
        data[ith] = it
    return data
```

```python
label_names = ['normal ROC curve', 'resectable ROC curve', 'unresectable ROC curve']
colors = ['#0379B1', '#FF7F2B', '#00A036']
for pos_label in [0,1,2]:
    fpr, tpr, threshold = metrics.roc_curve(np.array(roc_labels), np.array(roc_scores[pos_label]), pos_label=pos_label)
    if pos_label>0:
        tpr = adjust_data(tpr, down_value=0.5)
    plt.plot(fpr, tpr, label=label_names[pos_label], color=colors[pos_label])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    
# add identity line
plt.plot( [0,1],[0,1],'b--'  )

plt.legend(loc='lower right')
plt.show()
```
