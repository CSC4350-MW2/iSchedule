```python
import pandas as pd
import os

df = pd.read_csv("/Users/weizhenliu/Downloads/account_log_without_labels.csv")
df.groupby('ip_address').count().sort_values(['created_at'], ascending=False).iloc[:10, :]
```

We could see that there are two IP that has 100 or 35 "created_at account name	location	email". It's very strange as same IP can't has 35 accounts.

```python
df['malicious_account'] = (df['ip_address'] == "24.176.0.93") | (df['ip_address'] == "192.188.100.27")
```

```python
campaign = []
for i in df['ip_address']:
    if i == "24.176.0.93":
        campaign.append("cam_1")
    elif i == "192.188.100.27":
        campaign.append("cam_2")
    else:
        campaign.append("none")
df['campaign'] = campaign
```

```python
df.sort_values(['malicious_account'], ascending=False)
```
